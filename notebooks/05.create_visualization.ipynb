**Visualization 1**

idea: 
- use 4 colors to denote the actions, use linewidth to indicate the T(s'|s,a)
- only show visible if T > 0.3
- Show as tiles, each row is the transition from one cluster to others

```python
visual_save_dir = "../results/visualization"
version_date = "12012021"
T_mat = np.load(f'../results/intermediate/tmat_hierarchical_large_only_{version_date}.npy')
```

```python
def draw_tiles(T, vis_thresh=0.5):
    '''
    Draw the transition probability from source cluster to destination cluster after taking given actions.
    
    Arguments
    ---------------
    T: transition matrix. shape = [k, 4, k] where k is number of clusters. 4 is number of actions
    '''
    cmap = {
        0: "grey",
        1: "gold",
        2: "blue",
        3: "green"
    } 
    
    sides = ["up", "down"]
    
    custom_lines = [Line2D([0], [0], color='grey', lw=4),
                Line2D([0], [0], color='gold', lw=4),
                Line2D([0], [0], color='blue', lw=4),
                Line2D([0], [0], color='green', lw=4)]
    
    rect_w, rect_h = .18, .1

    def draw_rectangle(ax, name, position, w=rect_w, h=rect_h):
        # Draw a single rectangle at speicified position
        x, y = position
        rect = patches.Rectangle((x,y), w, h, linewidth=1, edgecolor='r', facecolor='blue', alpha=0.7)
        cx, cy = x + w/2, y + h/2
        ax.add_artist(rect)
        ax.annotate(name, (cx, cy), color='w', weight='bold', 
                    fontsize=14, ha='center', va='center')
        
    def draw_arror(ax, start_pos, end_pos, action, proba, camp):
        side = 'up' if action%2 == 0 else 'down'
        order = action // 2
        style = "Simple,tail_width=0.5,head_width=8,head_length=8"
        if side == 'up':
            cstyle = f"arc3,rad={-.2 - .15 * action}"
            start_pos = (start_pos[0], start_pos[1] + rect_h/2)
            end_pos = (end_pos[0], end_pos[1] + rect_h/2)
            txt_pos = (end_pos[0], end_pos[1] + .05 * (1 + order)) #avoid text overlap
        else:
            cstyle = f"arc3,rad={.2 + .15 * action}"
            start_pos = (start_pos[0], start_pos[1] - rect_h/2)
            end_pos = (end_pos[0], end_pos[1] - rect_h/2)
            txt_pos = (end_pos[0], end_pos[1] - .05 * (1 + order))
        kwargs = dict(arrowstyle=style, linewidth=proba*10, color=cmap[action], alpha=0.6)
        arrow = patches.FancyArrowPatch(start_pos, end_pos, connectionstyle=cstyle, **kwargs)
        ax.add_artist(arrow)
        ax.annotate(f'{round(proba,2)}', (txt_pos[0], txt_pos[1]), color='black', weight='bold',
                    fontsize=14, ha='center', va='center')

    s, a, s_prime = T.shape
    # Iterate through every starting state
    for i in range(s-2):
        fig = plt.figure(figsize=(15,5))
        ax = fig.add_subplot(111)
        heavy_trans_cluster = []
        cluster_position = {}
        for j in range(a):
            dest_group_idx = np.where(T[i, j, :] > vis_thresh)[0]
            dest_group_proba = T[i, j, :][dest_group_idx]
            for m, n in zip(dest_group_idx, dest_group_proba):
                heavy_trans_cluster.append([j, m, n]) # Action, destination, proba
        all_dest_group = list(set([k[1] for k in heavy_trans_cluster]))
        
        # Draw source rectangle on the left, and destination rectangles on the right
        draw_rectangle(ax, f'Decison Region {i}', (0.02, 0.4))
        start_pos = (0.02 + rect_w/2, 0.4 + rect_h/2)
        for j, c in enumerate(all_dest_group):
            if c == 15: name = 'Alive'
            elif c == 16: name = 'Death'
            else: name = f'Decison Region {c}'
            draw_rectangle(ax, name, (0.21 + j*0.19, 0.4))
            cluster_position[name] = (0.21 + j*0.19 + rect_w/2, 0.4 + rect_h/2)
        
        # Draw lines to connect source with destination
        for action, dest, proba in heavy_trans_cluster:
            if dest == 15:name = 'Alive'
            elif dest == 16: name = 'Death'
            else: name = f'Decison Region {dest}'
            end_pos = cluster_position[name]
            draw_arror(ax, start_pos, end_pos, action, proba, cmap)
        
        ax.legend(custom_lines, ['No Treatment', 'Fluid Only', 'Vaso Only', 'Fluid&Vaso'])
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        #fig.axes.append(ax)
        plt.savefig(f'{visual_save_dir}/Transition After Action - Cluster {i}', bbox_inches="tight")
        plt.show()
```

```python
draw_tiles(T_mat, vis_thresh=0.2)
```

**Visualization 2**

idea:

- x-axis: different features
- y-axis: mean values for each group

```python
def get_cluster_mean(column_names, cluster_arr_s1, cluster_arr_s2, A):
    '''
    For each clsuter, get feature means by taking different actions
    
    Arguments
    ---------------
    cluster_arr_s1: array of points belong to one cluster before taking action, 
                 shape = [N, f] where N is number of points, f is number of features. 
    A: actions every point at cluster take. Can be array of [N,] or [N, 4]
    cluster_arr_s2: array of points belong to other clusters after taking action,
                 shape = [N, f] where N is number of points, f is number of features. 
                 
    Output
    ---------------
    mean_by_feat: a dictionary. 
                  Key - name/index of each feature, 
                  Value - [mean at cluster, changed mean after action 0, changed mean after action 1,
                           changed mean after action 2, changed mean after action 3]
                  
    '''
    s1_means = np.mean(cluster_arr_s1, axis=0)
    cluster_means = {}
    offset = 0
    for feat_num, feat in enumerate(column_names):
        if feat == 'lactate':
            offset = 2
        else:
            offset = 0
        cluster_means[feat_num] = {}
        cluster_means[feat_num][0] = s1_means[feat_num]
        for a_num in range(len(np.unique(A))):
            idx = np.where(A==a_num)[0] + offset
            if offset == 0:
                changed_cluster_s2 = cluster_arr_s2[idx]
            else:
                changed_cluster_s2 = cluster_arr_s2[idx[:-offset]]
            cluster_means[feat_num][a_num+1] = np.mean(changed_cluster_s2[:, feat_num])
    return cluster_means

def draw_feat_mean_by_cluster(all_cluster_means, feat_names, version_date):
    '''
    Draw means for each feature by cluster
    
    Arguments
    ---------------
    all_cluster_means: a dictionary. 
                       Key - cluster number
                       Value - a dictionary. Output from get_cluster_mean function.
    '''
    num_of_cluster = len(all_cluster_means.keys())
    num_of_feats = len(list(all_cluster_means.values())[0].keys())
    custom_legend =[Line2D([0], [0], marker='o', markerfacecolor='black', color='w', label='Scatter', markersize=10),
                    Line2D([0], [0], marker='o', markerfacecolor='grey', color='w', label='Scatter', markersize=10),
                    Line2D([0], [0], marker='o', markerfacecolor='gold', color='w', label='Scatter', markersize=10),
                    Line2D([0], [0], marker='o', markerfacecolor='blue', color='w', label='Scatter', markersize=10),
                    Line2D([0], [0], marker='o', markerfacecolor='green', color='w', label='Scatter', markersize=10),
                   ]
    
    for i in range(num_of_feats):
        fig, ax = plt.subplots(1, 1, figsize=(15,5))
        for j in range(num_of_cluster):
            x = [j] * len(list(all_cluster_means[j][0].keys()))
            y = list(all_cluster_means[j][i].values())
            colors = ['black', 'grey', 'gold', 'blue', 'green']
            plot_colors = [colors[k] for k in all_cluster_means[j][0].keys()]
            ax.set_title(f'{feat_names[i]}', fontsize=22)
            ax.scatter(x, y, c=plot_colors, s=100, alpha=0.7)
            ax.set_xlabel('Decision Regions')
            ax.set_ylabel('Feature Value')
            ax.legend(custom_legend, 
                      ['Original', ' With No Treatment', 'After Fluid', 'After Vaso', 'After Fluid&Vaso'],
                      fontsize='x-large')
        plt.xticks(range(num_of_cluster))
        plt.savefig(f'{visual_save_dir}/Feature Mean Change by Cluster - {feat_names[i]} - {version_date}', bbox_inches="tight")
    plt.show()
```
